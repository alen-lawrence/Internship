```python
df.rename(columns={'Annual Income (k$)':'annual income'},inplace=True)
```

```python
df.rename(columns={'Spending Score (1-100)':'spending score'},inplace=True)
```

```python
df.head()
```

```python
#line plot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
x=df['CustomerID']
y=df['annual income']
sns.lineplot(x=x,y=y)
plt.title("asdf")
plt.xlabel("customer id")
plt.ylabel("annual income")
plt.show()
```

```python
#scatter plot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
x=df['Age']
y=df['spending score']
sns.scatterplot(x=x,y=y)
plt.title("asdf")
plt.xlabel("age")
plt.ylabel("spending score")
plt.show()
```

```python
#histogram
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
x=df['spending score']
sns.histplot(x=x,kde=30,bins=20)
plt.show()
```

```python
#box plot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(x=df['Genre'],y=df['annual income'],palette='dark')
plt.show()
```

```python
#bar plot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
x=df['annual income']
y=df['Age']
sns.barplot(x=x,y=y,palette='dark')
plt.show()
```

```python
#heatmap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#df=df.drop('Genre',axis=1)

sns.heatmap(df.corr(),annot=True)
plt.show()
```
